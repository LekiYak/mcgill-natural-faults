```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image
```

```python
im = Image.open('C:/Users/lekim/OneDrive/Desktop/School/research-2023/mcgill-natural-faults/surfaces/JBZQ04/EXP1/JBZQ04_cannon_scan_dark.png')

data = np.array(im, dtype=float)
```

```python
def plot(data, title):
    plot.i += 1
    plt.subplot(2,2,plot.i)
    plt.imshow(data)
    plt.gray()
    plt.title(title)
plot.i = 0
```

```python
plot(data, 'Original')


rgb_kernel = np.ones((3, 3, 3))

# A highpass filter
single_chan_kernel = np.array(
    [
        [-1, -1 , -1],  # nopep8
        [-1 , 4, -1],  # nopep8
        [-1, -1 , -1],  # nopep8
    ]  # nopep8
)  # nopep8

rgb_kernel[:, :, 0] = single_chan_kernel
rgb_kernel[:, :, 1] = single_chan_kernel
rgb_kernel[:, :, 2] = single_chan_kernel

filtered_img = ndimage.convolve(data, rgb_kernel)

plot(filtered_img, 'Highpass Filtered')

plt.show()
```
